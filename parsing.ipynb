```python
from datetime import datetime
import time
from tabulate import tabulate


f=open("syslogpy")
lines = f.readlines()
parser = []
for line in lines:
    case = []
#    print(line.split()[2])
    case.append(line.split()[0:3])
    case.append(line.split()[3])
    case.append(line.split()[4])
    case.append(line.split()[5:])
    parser.append(case)




today=datetime.today()
year=today.strftime("%Y")
#année non disponible dans le log, donc affichage de l'année actuel dans variable str

#print(parser)

#now = datetime.fromtimestamp(207578)
#print(now)
#year = 2022

for index in range(len(parser)):
    date=parser[index][0]
    date.insert(0, year)
    #insertion de l'année au début de la liste [2022/Sep/20,15:40:22]
    date=' '.join(date)
    #transformation de la liste en string "2022/Sep/20 15:40:22"
    timestp=time.mktime(datetime.strptime(date,"%Y %b %d %H:%M:%S").timetuple())
    #transformation du temps non normée format (2022/Sep/20 15:40:22) en timestamp (1663681222) - commande strp avec acquisition des modes 
    dt_object = datetime.fromtimestamp(timestp).strftime('%d/%m/%Y %H:%M:%S')
    #transformation du timestamp (1663681222) en temps normée format (2022-09-20 15:40:22)

    #https://docs.python.org/3/library/datetime.html
    #https://www.programiz.com/python-programming/datetime/strftime
    #print(dt_object)

    parser[index][0] = dt_object
    #instertion dans la liste 


#print(parser)
#print(date)
#timestamp = datetime.timestamp(strparser)
#timestamp = time.mktime(datetime.strptime(strparser, "%d/%m/%y : %H:%M/%S").timetuple())

#création du fichier syslogpars depuis le tableau parser

#outputList = dictOfOutputs.items()
#table = outputList
with open('syslogpypars', 'w') as f:
    f.write(tabulate(parser))

#print(parser)

#supression de la premiere ligne

with open(r"syslogpypars", 'r+') as fp:
    lines = fp.readlines()
    fp.seek(0)
    fp.truncate()
    fp.writelines(lines[1:])
```

```python
from datetime import datetime
import time
import pandas 


f=open("dpkgpy")
lines = f.readlines()
parserdpkg = []
for line in lines:
    case = []
#    print(line.split()[2])
    case.append(line.split()[0:2])
    case.append(line.split()[2])
    case.append(line.split()[3:])
    #case.append(line.split()[5:])
    parserdpkg.append(case)





#print(parser)

#now = datetime.fromtimestamp(207578)
#print(now)
#year = 2022

for index in range(len(parserdpkg)):
    date2=parserdpkg[index][0]
    date2=' '.join(date2)
    timestp=time.mktime(datetime.strptime(date2,"%Y-%m-%d %H:%M:%S").timetuple())
    #transformation du temps non normée format (2022/Sep/20 15:40:22) en timestamp (1663681222) - commande strp avec acquisition des modes 
    dt_object = datetime.fromtimestamp(timestp).strftime('%d/%m/%Y %H:%M:%S')
    #transformation du timestamp (1663681222) en temps normée format (2022-09-20 15:40:22)

    #https://docs.python.org/3/library/datetime.html
    #https://www.programiz.com/python-programming/datetime/strftime
    #print(dt_object)

    parserdpkg[index][0] = dt_object
    #instertion dans la liste 

with open('dpkgpypars', 'w') as f:
    f.write(tabulate(parserdpkg))

#supression de la premiere ligne

with open(r"dpkgpypars", 'r+') as fp:
    lines = fp.readlines()
    fp.seek(0)
    fp.truncate()
    fp.writelines(lines[1:])

#data2=pandas.read_csv('parserdpkg',sep=";")
#a=pandas.merge(parser,parserdpkg)
#print(a)
```
